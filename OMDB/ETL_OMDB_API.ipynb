```python
titles=tom_hanks_df['title'].tolist()
titles
```

```python
url = "http://www.omdbapi.com/?apikey=" + api_key + "&t="
movies = titles
responses = []
for movie in movies:
    movie_data = requests.get(url + movie).json()
    responses.append(movie_data)
```

```python
hanks_movies=pd.DataFrame(responses)
```
